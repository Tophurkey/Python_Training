```python
stocks = ['AMC','GME','AAPL','MSFT','TSLA']

#create a function to pull stock data from api
def _stock_pull(ticker):
    querystring = {"function":"TIME_SERIES_DAILY","symbol":ticker,"outputsize":"compact","datatype":"json"}

    response1 = requests.request("GET", url, headers=headers, params=querystring)

    #stores the response json into a dictionary
    loop_dict = response1.json()
    #select only the daily data and not the metadata fromt he dictionary
    daily_loop = loop_dict['Time Series (Daily)']
    #create a dataframe fromt the dictionary
    df_loop = pd.DataFrame(daily_loop)
    #Transpose the dataframe
    df_loop_t = df_loop.transpose()
    #rename the columns
    df_loop_t.columns = ['open','high','low','close','volume']
    #save the dataframe to an excel file
    df_loop_t.to_excel(f'{ticker}_daily.xlsx')
```

```python
#loop through the stock pulling function to get all the data
for ticker in stock:
    _stock_pull(ticker)

#pull a single stock by ticker
_stock_pull('ILMN')
```
